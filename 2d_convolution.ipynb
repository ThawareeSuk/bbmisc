from http://docs.scipy.org/doc/scipy-0.15.1/reference/generated/scipy.signal.convolve2d.html

```python
from scipy import signal
from scipy import misc
import numpy as np
n,L,r = 201,2,1
v = np.linspace(-L,L,n,endpoint=True)
x = np.tile(v,(n,1))
y = np.tile(v[:,np.newaxis],(1,n))
tophat = (x**2+y**2<r**2).astype(float)
c = signal.convolve2d(tophat,tophat)
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
fig, (ax_top, ax_conv) = plt.subplots(1, 2)
ax_top.imshow(tophat,extent=[-L,L,-L,L])
ax_conv.imshow(c,extent=[-L,L,-L,L])
plt.show()
```

```python
fig, ax = plt.subplots()
ax.plot(c[n,:]);
```

I *think* the slight curvature near the intersection is a numerical artifact.

```python
from mpl_toolkits.mplot3d import Axes3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
nn = 2*n-1
v2 = np.linspace(-L,L,nn,endpoint=True)
x2 = np.tile(v2,(nn,1))
y2 = np.tile(v2[:,np.newaxis],(1,nn))
ax.plot_surface(x2,y2,c);
```
